```python
from conllu import parse_incr
import pandas as pd

def GetSentences(filepath, desired_length):
    sentences = []
    with open(filepath, "r", encoding="utf-8") as f:
        for tokenlist in parse_incr(f):
            sentences.append(tokenlist)
    initial_length = len(sentences)
    sentences = sentences[:desired_length] if initial_length > desired_length else sentences
    print(f"Number of sentences loaded: {len(sentences)} (original length: {initial_length})")
    return sentences
```

```python
def GetPrompt(prompt_path):
    with open(prompt_path, "r", encoding="utf-8") as f:
        return f.read()
```

```python
def TestTreebank(sentences, prompt, output_csv="results.csv"):
    rows = []
    for i, sentence in  enumerate(sentences):
        print(f"Обрабатывается предложение {i+1}/{len(sentences)}...") #we will track the processing to find bugs
        core_concepts = client.chat.completions.create(messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": f"Sentence: {sentence}"}
        ],
        # Форматирование ответа
        #response_format=
        #{
        #    "type": "json_object"
        #},
        model='llama-3.1-8b-instant', # Тип модели
        stream=False,
        max_tokens=8192, # Максимальная длина выдачи, желательно поменять под задачу
        temperature=0.7, # Температура
        top_p=0.5 # Топ-P отвечает за разнообразие выдачи
        ).choices[0].message.content
        rows.append({"sentence": sentence, "response": core_concepts})
    df = pd.DataFrame(rows)
    df.to_csv(output_csv, index=False, encoding="utf-8-sig")
```
